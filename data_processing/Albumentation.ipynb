```python
import json
import numpy as np
import pandas as pd
import cv2
import albumentations as A
from matplotlib import pyplot as plt
from albumentations.pytorch import ToTensorV2
```

```python
def show_img(img):

    plt.figure(figsize=(6,6))
    plt.axis('off')
    
    plt.imshow(img)
```

```python
cow = cv2.imread('cow.jpg')
cow = cv2.cvtColor(cow, cv2.COLOR_BGR2RGB) 
```

```python
show_img(cow)
```

```python
transform = A.Blur(blur_limit=(3, 30), p=1)
blurcow = transform(image=cow)['image']
show_img(blurcow)
```

```python
transform = A.GlassBlur(max_delta=12, iterations=2, p=1)
blurcow = transform(image=cow)['image']
show_img(blurcow)
```
